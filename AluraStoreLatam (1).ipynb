#1. Análisis de facturación



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

ingreso_tienda1 = tienda1["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print("Ingresos por tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt

ingresos = [
    tienda1["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

plt.figure(figsize=(8, 5))
plt.bar(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], ingresos, color='skyblue')
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]

for i, tienda in enumerate(tiendas, start=1):
    productos = tienda["Producto"].value_counts()
    top = productos.head(5)
    bottom = productos.tail(5)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    fig.suptitle(f'Tienda {i}: Productos Más y Menos Vendidos', fontsize=14)

    axes[0].bar(top.index, top.values, color='green')
    axes[0].set_title('Más vendidos')
    axes[0].set_ylabel('Cantidad vendida')
    axes[0].tick_params(axis='x', rotation=45)

    axes[1].bar(bottom.index, bottom.values, color='red')
    axes[1].set_title('Menos vendidos')
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

# 5. Envío promedio por tienda

```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

envio_t1 = tienda1["Costo de envío"].mean()
envio_t2 = tienda2["Costo de envío"].mean()
envio_t3 = tienda3["Costo de envío"].mean()
envio_t4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio por tienda:")
print(f"Tienda 1: ${envio_t1:,.2f}")
print(f"Tienda 2: ${envio_t2:,.2f}")
print(f"Tienda 3: ${envio_t3:,.2f}")
print(f"Tienda 4: ${envio_t4:,.2f}")
```

```python

calificaciones = [
    tienda1["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

costos_envio = [
    tienda1["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

plt.figure(figsize=(8, 5))
plt.plot(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], calificaciones, label="Calificación Promedio", marker='o')
plt.plot(["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"], costos_envio, label="Costo de Envío Promedio", marker='s')
plt.title("Calificación vs. Costo de Envío por Tienda")
plt.ylabel("Valor promedio")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

El objetivo de este análisis fue determinar cuál de las cuatro tiendas disponibles representa la mejor oportunidad para que el Sr. Juan comience a vender sus productos. Para ello, se evaluaron varios factores clave, incluyendo los ingresos totales generados por cada tienda, las categorías de productos más y menos vendidas, las calificaciones promedio otorgadas por los clientes, los productos destacados en términos de ventas y el costo de envío promedio. A partir de estos análisis, se generaron visualizaciones que ayudaron a identificar patrones relevantes y facilitar la toma de decisiones.

En primer lugar, se analizaron los ingresos totales por tienda. Si bien algunas tiendas demostraron mayor volumen de ventas que otras, fue importante considerar que el ingreso por sí solo no garantiza una buena experiencia de venta para el Sr. Juan. Algunas tiendas con altos ingresos también presentaron gran competencia interna o costos operativos mayores, lo cual podría afectar sus márgenes de ganancia o la visibilidad de sus productos.

Respecto a las categorías de productos, se observó que cada tienda tiene una distribución distinta en cuanto a los tipos de productos más vendidos. Identificar si el tipo de producto que el Sr. Juan planea vender coincide con las categorías más populares de una tienda fue crucial para aumentar sus probabilidades de éxito. Tiendas con fuerte demanda en la misma categoría de su producto representan una oportunidad, siempre que no estén saturadas de competencia directa.

En cuanto a la satisfacción del cliente, las calificaciones promedio otorgadas a los productos ofrecieron una valiosa perspectiva sobre la calidad de la experiencia de compra en cada tienda. Se observó que una de las tiendas destaca notablemente con la calificación promedio más alta, lo cual sugiere un entorno favorable para nuevos vendedores interesados en ofrecer un buen servicio.

Al analizar los productos más y menos vendidos, se encontró que ciertas tiendas concentran sus ventas en artículos que están directamente relacionados con lo que el Sr. Juan planea ofrecer. Este es un indicador claro de afinidad con la demanda existente. También se evaluó la diversidad de productos, ya que en algunas tiendas las ventas están muy concentradas en pocos artículos, lo que puede limitar la visibilidad de nuevos productos.

Por último, se evaluó el costo promedio de envío en cada tienda. Este aspecto es especialmente importante porque puede influir en la decisión final de los compradores. Las tiendas con costos de envío más bajos tienden a facilitar más ventas, ya que reducen la barrera económica para el cliente. En este sentido, se identificó una tienda con un costo de envío significativamente menor, lo que representa una ventaja competitiva.

Con base en todos estos factores, se concluye que la Tienda 3 representa la mejor opción para que el Sr. Juan comience a vender. Esta tienda no solo tiene la calificación promedio más alta, lo que refleja una excelente percepción de los clientes, sino que también cuenta con el menor costo de envío, lo cual puede aumentar la conversión de ventas. Además, los productos más vendidos en esta tienda están alineados con la oferta del Sr. Juan, lo que indica una demanda real en esa categoría. Aunque otras tiendas puedan presentar ingresos ligeramente mayores, la Tienda 3 ofrece un equilibrio ideal entre satisfacción del cliente, eficiencia logística y afinidad con el tipo de producto, por lo tanto, es la opción más estratégica para el inicio de su negocio.
